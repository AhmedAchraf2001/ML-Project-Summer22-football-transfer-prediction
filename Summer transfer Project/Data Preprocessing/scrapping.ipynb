```python
name = olddata.name[(olddata[olddata.name == ''].index)+1:(olddata[olddata.name == '' ].index)+10000]
name
```

```python
ffx = webdriver.Chrome()

ffx.get("https://www.transfermarkt.com")


bigLi = []
counter = 0

for e in name:
    #search
    li = []
    counter+=1
    li.append(e)
    ffx.find_element(By.CSS_SELECTOR, '[class="tm-header__input--search-field"]').send_keys(e) # Search with player name
    ffx.find_element(By.CSS_SELECTOR,'[class="tm-header__input--search-send"]').send_keys(Keys.ENTER) # press Enter Key

    
    player = ffx.find_element(By.CSS_SELECTOR,'td[class= "hauptlink"]')
    url = player.find_element(By.TAG_NAME,'a').get_attribute('href')

    for st in ['leistungsdatenverein', 'nationalmannschaft', 'marktwertverlauf']:
        x = url.split('/')
        x[4] = st
        for i in range(len(x)):
            x[i] = x[i] + "/"
        url = ''.join(str(u) for u in x)

        ffx.get(url)
        
        try:
            if st != 'marktwertverlauf':  
                stat = ffx.find_element(By.CSS_SELECTOR, 'div[id="yw1"]').find_element(By.TAG_NAME,'tfoot')

                for i, j in stat.find_elements(By.CSS_SELECTOR, 'td[class="zentriert"]'), range(1,9):
                    li.insert(j, i.get_attribute("innerHTML"))

                for x, j in stat.find_elements(By.CSS_SELECTOR, 'td[class="rechts"]'), range(9,17):
                    li.insert(j, x.get_attribute("innerHTML"))

            else:
                for i, j in ffx.find_elements(By.CSS_SELECTOR, 'span[class="quick-fact__content--large"]'), range(17,23):
                    li.insert(j,i.get_attribute("innerHTML"))
        
        except :
            for m in range(0,22):
                 li.append('none')

    bigLi.append(li)
                
                
ffx.quit()
```
